```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Viz avec Pandas

#### plot y en fonction de x

```python
 A = np.array(range(1, 1000 + 1))
 B = np.log(A)
 C = np.random.randint(2, size = 1000)
df = pd.DataFrame(data= {"A": A, "B": B, "C": C}, index= range(1000))
df.head(5)
```

```python
df.plot(x = 'A', y= 'B')
```

#### histogramme

```python
A= np.random.normal(0, 1, 1000000)
df = pd.DataFrame({"A": A})
```

```python
df.hist('A', bins = 1000)
```

#### boxplot

```python
# charger le dataset 
titanic =  pd.read_csv('titanic.csv')
titanic.head(3)
```

```python
titanic.boxplot('Age')
```

#### pie

```python
titanic[['Sex', 'Survived']].groupby(['Sex']).sum().plot(y= 'Survived', kind='pie', colors= ['red', 'blue'], autopct='%1.0f%%',  title='survécus de Titanic selon le sex' )
```

#### bar

```python
titanic.groupby('Sex')['Age'].mean().plot(x = 'Sex', y= 'Age', kind= 'bar', title= "Moyenne d'âge selon le sexe")
```

#### correlation matrix 

```python
corrmat = titanic.corr()
```

```python
plt.figure(figsize=(13, 6))
sns.heatmap(corrmat, annot=True)
plt.xticks(rotation=30)
plt.show()
```
